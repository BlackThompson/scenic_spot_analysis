```python
import os
import requests


def download_file(url, download_path, file_name):
    # ç¡®ä¿ä¸‹è½½ç›®å½•å­˜åœ¨ï¼Œå¦‚æœä¸å­˜åœ¨åˆ™åˆ›å»ºå®ƒ
    if not os.path.exists(download_path):
        os.makedirs(download_path)

    # æ„å»ºå®Œæ•´çš„æ–‡ä»¶ä¿å­˜è·¯å¾„
    file_path = os.path.join(download_path, file_name)
    
    # å¦‚æœæ–‡ä»¶å·²ç»å­˜åœ¨ï¼Œåˆ™è·³è¿‡ä¸‹è½½
    if os.path.exists(file_path):
        print(f"æ–‡ä»¶ {file_path} å·²å­˜åœ¨ï¼Œè·³è¿‡ä¸‹è½½")
        return

    # å‘é€HTTP GETè¯·æ±‚ä»¥ä¸‹è½½æ–‡ä»¶
    response = requests.get(url)

    # æ£€æŸ¥å“åº”çŠ¶æ€ç ï¼Œ200è¡¨ç¤ºæˆåŠŸ
    if response.status_code == 200:
        # ä»¥äºŒè¿›åˆ¶æ¨¡å¼å†™å…¥æ–‡ä»¶
        with open(file_path, "wb") as file:
            file.write(response.content)

        print(f"æ–‡ä»¶å·²ä¸‹è½½åˆ° {file_path}")
    else:
        print(f"ä¸‹è½½å¤±è´¥ï¼ŒçŠ¶æ€ç ï¼š{response.status_code}")


def download_picture(data_spot, picture_dir_path):
    image_list = data_spot["image_list"]
    for i in range(len(image_list)):
        # æ–°å»ºæ–‡ä»¶å¤¹
        picture_dir_path_each = picture_dir_path + "/" + str(i)
        # print(picture_dir_path_each)
        # if not os.path.exists(picture_dir_path_each):
        #     os.makedirs(picture_dir_path_each)
        each_image_list = image_list[i].split(",")
        for j in range(len(each_image_list)):
            download_url = each_image_list[j]
            download_file(download_url, picture_dir_path_each, str(j) + ".jpg")
```

```python
import os
import pandas as pd

# éå†E:\Python\Code\scenic_spot_analysis\data\spot_excelæ–‡ä»¶å¤¹ä¸‹çš„æ‰€æœ‰æ–‡ä»¶
excel_path = r"./data/spot_excel/"
picture_dir = r"./data/picture/"
file_list = os.listdir(excel_path)
for file_name in file_list:
    name = file_name.split(".")[0]
    file_path = excel_path + file_name
    picture_dir_path = picture_dir + name

    data_spot = pd.read_csv(file_path)
    download_picture(data_spot, picture_dir_path)
```
